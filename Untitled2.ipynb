```python
seoul =pd.read_csv('/content/sample_data/서울기온.csv')
seoul
```

```python
seoul = seoul.drop(seoul.index[0:6])
seoul
```

```python
seoul.columns
```

```python
seoul.columns = ['날짜','지점','평균기온','최저기온','최고기온']
seoul.info()
```

```python
seoul = seoul.drop('지점', axis=1)
seoul
```

```python
seoul.isnull().sum()
```

```python
seoul = seoul.dropna()
```

```python
seoul.isnull().sum()
```

```python
seoul['날짜'] = seoul['날짜'].str.replace('\t','')
seoul
```

```python
seoul['년도'] = seoul['날짜'].str[:4]
seoul
```

```python
seoul0815 = seoul[seoul['날짜'].str.contains('08-15')]
seoul0815
```

```python
fig = plt.figure(figsize=(15,7))

x =seoul0815['년도'].astype(int)
y =seoul0815['평균기온'].astype(float).sort_values()

plt.xlabel('years')
plt.ylabel('meantemp')
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

#2024년 8월15일 기온 예측하기

x = seoul0815['년도'].astype(np.int64).values.reshape(-1, 1) # x를 2차원 배열로 변환
y = seoul0815['평균기온'].astype(float).values # Series에서 값 추출

model.fit(x, y)
result = model.predict([[2024]])
print('예상기온:',result)
```
